```python
files_ = ['Fechamento_Tap_rev01.xlsx','Fechamento_Ara_rev01.xlsx',
         'Fechamento_Cat_rev01.xlsx','Fechamento_Pat_rev01.xlsx',
         'Fechamento_Caj_rev01.xlsx'
         ]
idx = 4
tapira =['Barragem BR', 'Barragem BL1', 'Barragem BD2', 'Barragem BD5', 'Barragem BA3']
araxa = ['Barragem B1/B4', 'Barragem B5', 'Barragem B2','Barragem B6']
catalao =['Barragem de Magnetita', 'Barragem de Rejeitos BR']
patrocinio = ['Barragem de Lamas', 'Barragem de Lamas (cenário 1)']
cajati = ['Barragem B1', 'Barragem B2', 'Barragem B12']
barragens_check = [tapira, araxa, catalao, patrocinio, cajati]
barragens_names = ['tapira', 'araxa', 'catalao', 'patrocinio', 'cajati']
```

type(nomes)

```python
col_list_names = []
row_list_names = []
for idx in range(len(barragens_check)):
    barragens =[]
    rows =[]
    cont = -1
    df = pd.read_excel(files_[idx],skiprows=3, header=None)
    nomes = list(df.iloc[:,2].values) # recovery names of structures
    nrows,ncols = df.shape
    
    cols_drop = [] # find colunms without any value
    for idx_col in range(ncols):
        num_na = df.iloc[:,idx_col].isna().sum()
        if num_na == nrows:
            cols_drop.append(idx_col)
            
    df = df.drop(cols_drop,axis = 1)  #drop columns without any value 
    
    for word in nomes:
        cont += 1
        try:
            if word in barragens_check[idx]:
                barragens.append(word)
                row_list_names.append(barragens_names[idx]+ ' ' +word)
                rows.append(cont)
        except:
            pass
    df_filter = df.iloc[rows,:]
    print(df.shape, df_filter.shape,barragens_check[idx],barragens_names[idx])
    
    cols_names = []
    
    for word in df.iloc[1,:].values:  #recovery the name of columns
        if isinstance(word, str):
            cols_names.append(word)
    col_list_names.append(cols_names)
    

    
    
```
