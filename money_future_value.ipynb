```python
import matplotlib.pyplot as plt
```

```python
period = 10 # years
FV= 0 # future value
i = 0.92 # 1% per month =12.7% per year
Dmin=1000 # deposit by month
Dmax =3000 # max value deposit by month
delta = 100 # 10 # linear increase per month
```

```python
stop = period*12
```

```python
time = 0
FV= 0 # future value
D = Dmin - delta
soma,Dsoma, t,money,Dep = 0,[],[],[], []

while time < stop:
  D = D + delta
  time = time + 1
  if D > Dmax:
    D = Dmax
  soma = soma +D
  FV = FV*(1+i/100) + D
  money.append(FV)
  t.append(time)
  Dep.append(D)
  Dsoma.append(soma)
```

```python
plt.plot(t,money)
plt.plot(t,Dsoma)
plt.xlabel('tempo-meses')
plt.ylabel('valor acumulado')
plt.legend(['Total', 'Valor depositado'])
```

```python
plt.plot(t,Dep)
plt.xlabel('tempo-meses')
plt.ylabel('valor investido no mes')
```

```python
time = 0
stop = 50 # number of years you will recive money
fator = 70*12 # valor americano de referencia 70 vezes o gasto anual
# vc pode trocar por 200 ou simplesmente fixar o valor fixo de PV
# compo por exemplo fazer PV = 1000000
Salary = 5000
PV = fator* Salary

soma,spend,available,t = 0,[],[], []
i = 13 # interest rate per year


i = (( i/100 + 1))**(1/12) -1 # interest rate converting to rate by month


stop = stop*12 # time convertig to months
while (time < stop) or (PV > 0):
  soma = soma + Salary
  time = time + 1
  PV = (PV-Salary)*(1+i/100)
  if PV < 0:
    break
  spend.append(soma/1000)
  t.append(time/12)
  available.append(PV/1000)
```

```python
n = len(t)-1
print(available[n],t[n])
```

```python
plt.plot(t,spend)
plt.plot(t,available)
plt.xlabel('tempo-anos')
plt.ylabel('valor acumulado/1000 ')
plt.legend(['gasto', 'disponivel'])
```
