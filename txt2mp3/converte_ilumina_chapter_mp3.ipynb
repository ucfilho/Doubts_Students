```python
# Import the required module for text 
# to speech conversion
import pyttsx3
import unicodedata #no need to install, it is default library in python
```

```python
#!pip install pyttsx3
```

```python
# name of the file
file_name = 'A Verdadeira Essência da Iluminação Osho -parte 01' # Name or path to your text file

file_path = file_name + '.txt'
```

```python
# read the text

try:
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
        #print(content)
except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
def normalize_text(text):
    return unicodedata.normalize("NFKC", text)

mytext = normalize_text(content)
```

```python
engine = pyttsx3.init()
engine.setProperty('rate', 300)  # speed
engine.setProperty('volume', 1.0)

mp3 = file_name + '.mp3'

engine.save_to_file(mytext, mp3)
engine.runAndWait()
```
